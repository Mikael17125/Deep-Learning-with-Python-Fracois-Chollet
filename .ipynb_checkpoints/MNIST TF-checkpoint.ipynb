```python
from tensorflow.keras.datasets import mnist
```

```python
import numpy as np
from tensorflow import keras 
from tensorflow.keras import layers
```

```python
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

```python
train_images.shape
```

THIS MEANS I CREATED MATRIX WITH SHAPE (INPUT, 512) => ACTUALLY THIS IS SAME WITH WEIGTH SHAPE
AND 2ND LAYER MEAN I CREATED MATRIX WITH SHAPE (LAST-LAYER, 10) AND ACTIVATION FUNCTION IS USED FOR CONVERT THE NUMBER TO BE SOME RANGE

```python
model = keras.Sequential([
    layers.Dense(512, activation = "relu"),
    layers.Dense(10, activation = "softmax")
])
```

OPTIMIZER IS USED FOR CALCULATE THE WEIGHT, I JUST KNOW ONE OF THE OPTIMIZER IS GRADIENT DECENT
LOSS IS USED FOR CALCULATED HOW FAR THE PREDICTION FROM THE ACTUAL VALUE

```python
model.compile(optimizer="rmsprop",
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"])
```

STILL DON'T GET IT WHY THE PIXEL IS NORMALIZED

```python
train_images = train_images.reshape((60000, 28 * 28))
train_images = train_images.astype("float32") / 255 
test_images = test_images.reshape((10000, 28 * 28))
test_images = test_images.astype("float32") / 255
```

FIT MODEL IS USED FOR TRAIN MODEL WITH SOME EPOCH AND BATCH SIZE
AS I KNOW BATCH SIZE IS LIKE NUMBER STEP TO NEXT DATA TRAIN SO THE DATA TRAIN ACTUALLY SURELY NOT ARRANGED

THE MATH FOR BATCH SIZE IS:
    ONE EPOCH EQUALS TO TOTAL DATA TRAIN DIVIDED BY BATCH SIZE
    
EXAMPLE :
DATA TRAIN FOR THIS PROJECT IS EQUAL TO 60.000 
AND BATCH SIZE TUNED IS 64
SO ONE EPOCH EQUALS TO 60.000 / 64 = 937.5 AND CEIL = 938

```python
model.fit(train_images, train_labels, epochs=10, batch_size=64)
```

```python
test_digits = test_images[0:10]
predictions = model.predict(test_digits)
```
